```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Load and clean data
data = pd.read_csv("output_mpi.csv", header=None)
data = data.apply(pd.to_numeric, errors="coerce").dropna(axis=1)  # Drop non-numeric columns
data_transposed = data.T

# Define x and y range
shape0, shape1 = data_transposed.shape
x = np.linspace(-3, 3, shape1)
y = np.linspace(0, 4, shape0)

# Plot 3D surface
fig = go.Figure(data=[go.Surface(z=data_transposed.values, x=x, y=y)])
fig.update_layout(
    title="Function w(x,y) - Transposed",
    autosize=False,
    width=800,
    height=800,
    scene=dict(
        xaxis_title="Y Axis",
        yaxis_title="X Axis",
        zaxis_title="Z Axis",
    ),
)
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Load and clean data
data = pd.read_csv("output.csv", header=None)
data = data.apply(pd.to_numeric, errors="coerce").dropna(axis=1)  # Drop non-numeric columns
data_transposed = data.T

# Define x and y range
shape0, shape1 = data_transposed.shape
x = np.linspace(-3, 3, shape1)
y = np.linspace(0, 4, shape0)

# Plot 3D surface
fig = go.Figure(data=[go.Surface(z=data_transposed.values, x=x, y=y)])
fig.update_layout(
    title="Function w(x,y) - Transposed",
    autosize=False,
    width=800,
    height=800,
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis",
    ),
)
fig.show()
```

```python
import matplotlib.pyplot as plt

# 
omp_threads = [1, 2, 4, 8]
acceleration_1 = [1.000, 0.689, 0.575, 0.555] # 
acceleration_2 = [1.000, 0.674, 0.543, 0.489] # 

# 
plt.figure(figsize=(8, 6))
plt.plot(omp_threads, acceleration_1, marker='o', linestyle='-', color='b', label='80*90')
plt.plot(omp_threads, acceleration_2, marker='o', linestyle='-', color='r', label='160*180')

# 
plt.title('Acceleration vs OpenMP Threads', fontsize=14)
plt.xlabel(' OpenMP-   (Number of OpenMP Threads)', fontsize=12)
plt.ylabel(' (Acceleration)', fontsize=12)
plt.xticks([1, 2, 4, 8]) # 
plt.grid(True, linestyle='--', alpha=0.6)

# 
plt.legend(fontsize=12)

# 
plt.tight_layout()
plt.show()
```
